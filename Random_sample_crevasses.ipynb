```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import rasterio as rio
import os
```

```python
# replace path name to your directory that has all_tiles_april
tile_dir = "/home/data/all_tiles_april/"
file_list = os.listdir(tile_dir)
file_list.sort()
```

```python
tile = np.random.choice(file_list)
sample = rio.open(tile_dir + tile)
coords = sample.lnglat()
```

```python
plt.imshow(sample.read().reshape((200,200)))
plt.colorbar()
print(tile)
print(coords)
```
