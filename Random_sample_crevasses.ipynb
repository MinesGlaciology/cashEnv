```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
import os
from IPython.display import clear_output
```

```python
# replace path name to your directory that has all_tiles_april
tile_dir = "./output/"
file_list = os.listdir(tile_dir)
file_list.sort()
```

```python
clear_output()
tile = np.random.choice(file_list)
sample = rio.open(tile_dir + tile)
coords = sample.lnglat()

plt.imshow(sample.read().reshape((200,200)))
plt.colorbar()
plt.show()
print(tile)
print(coords)
```
