Basic notebook to test the performance of the quicksom library

```python
import torch
from quicksom.som import SOM
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
device
```

```python
X = numpy.load('/home/data/Full_DataDCT.npy')
X = torch.from_numpy(X[:1000])
X = X.float()
X = X.to(device)
```

```python
# Create SOM object and train it, then dump it as a pickle object
m, n = 100, 100
dim = X.shape[1]
niter = 50
batch_size = 100
som = SOM(m, n, dim, niter=niter, device=device)
learning_error = som.fit(X, batch_size=batch_size)
```
